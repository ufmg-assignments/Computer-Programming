```python
def exercicio_3(tupla):
    L = []
    A = B = ()
    for k in range (0,len(tupla)//2):
        A += tupla[k],
    L.append(A)
    for k in range (len(tupla)//2,len(tupla)):
        B += tupla[k],
    L.append(B)
    return L

tupla = (1,25,3,4,46,4,5,6,7)

exercicio_3(tupla)
```

```python
def exercicio_4(tupla):
    L = []
    A = B = ()
    for k in range (0,len(tupla)):
        if tupla[k] %2 == 0:
            A += tupla[k],
    L.append(A)
    for k in range (0,len(tupla)):
        if tupla[k] %2 == 1:
            B += tupla[k],
    L.append(B)
    return L

tupla = (1,222,2,3,3,4,5,6,7,12,91,15)
exercicio_4(tupla)
```
